## IPA Plots

```python
def truncate(number):
    factor = 10 ** 3
    return math.floor(number * factor) / factor

def printAllIPA(n): 
    list = []
    list.append(Pupil.from_pupil_capture("../../data/colet/participant" + str(n) +"/task0/pupil.csv"))
    list.append(Pupil.from_pupil_capture("../../data/colet/participant" + str(n) +"/task1/pupil.csv"))
    list.append(Pupil.from_pupil_capture("../../data/colet/participant" + str(n) +"/task2/pupil.csv"))
    list.append(Pupil.from_pupil_capture("../../data/colet/participant" + str(n) +"/task3/pupil.csv"))
    print("IPA data for participant " + str(n))
    print("Task 1: " + str(truncate(list[0].ipa())))
    print("Task 2: " + str(truncate(list[1].ipa())))
    print("Task 3: " + str(truncate(list[2].ipa())))
    print("Task 4: " + str(truncate(list[3].ipa())))



def plotAllIPA(n):
    ipa_values = []
    tasks = ["Task 1", "Task 2", "Task 3", "Task 4"]

    for i in range(4):
        pupil_data = Pupil.from_pupil_capture(f"../../data/colet/participant{n}/task{i}/pupil.csv")
        ipa = truncate(pupil_data.ipa())
        ipa_values.append(ipa)

    # Print IPA values
    print(f"IPA data for participant {n}")
    for i, val in enumerate(ipa_values):
        print(f"{tasks[i]}: {val}")

    # Plotting the bar chart
    plt.figure(figsize=(6, 4))
    plt.bar(tasks, ipa_values)
    plt.title(f"IPA Values for Participant {n}")
    plt.ylabel("Index of Pupillary Activity")
    plt.ylim(0, 1)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.show()

    # Calculate relative ratios
    baseline = ipa_values[0]
    ratio_list = [
        1,
        ipa_values[1] / baseline if baseline != 0 else 0,
        ipa_values[2] / baseline if baseline != 0 else 0,
        ipa_values[3] / baseline if baseline != 0 else 0,
    ]

    return ratio_list
```

```python
lst = [0, 0, 0, 0]
for i in range (30):
    temp = plotAllIPA(i)
    for j in range(4):
        lst[j] += temp[j]
    print("-----------------------------------------------------------------------------\n\n")
    
factor = lst[0]

for i in range (4):
    lst[i] /= factor

print("average ratios: " + str(lst))
```

## RIPA Plots

```python
def plotSegmentedRIPA(n, task, window: int = 360, step: int = 90):
    pupil_data = Pupil.from_pupil_capture(f"../../data/colet/participant{n}/task{task-1}/pupil.csv")
    segments = pupil_data.split_into_segments(window, step)
    
    ripaList = []
    start_times = []

    for pupil in segments:
        ripa = pupil.ripa()
        ripaList.append(ripa)
        start_times.append(pupil.t[0])  # start of each segment

    # Plot
    plt.figure(figsize=(16, 8))
    plt.plot(start_times, ripaList)
    plt.title(f"Segmented RIPA values - Participant {n}, Task {task}")
    plt.xlabel("Segment Start Time (s)")
    plt.ylabel("RIPA Value")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
for i in range(3):
    for j in range(4):
       plotSegmentedRIPA(i, j + 1) 
```
