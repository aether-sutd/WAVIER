```python
with open("../../data/colet/colet_v3.json") as f:
    text = f.read()
```

```python
lines = text.split("\n")
len(lines)
```

```python
lines[:6]
```

```python
start = None
idx = []
objects = []

for i, line in enumerate(lines):
    if line.startswith("  {"):
        start = i
    elif start is not None and line.startswith("  }"):
        idx.append((start, i+1))
        objects.append("\n".join(lines[start:i+1]).strip().strip(","))
        start = None

len(objects)
```

```python
for idx_start, idx_end in idx[::-1]:
    lines = lines[:idx_start] + lines[idx_end:]

len(lines)
```
