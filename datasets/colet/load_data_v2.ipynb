```python
import pandas as pd
from glob import glob
import os
```

```python
pd.read_csv("../../data/colet/exports/participant1_task1_blinks.csv")
```

```python
import json

for participant in range(1, 48):
    os.makedirs(f"../../data/colet/participant{participant-1}", exist_ok=True)
    for task in range(1, 5):
        os.makedirs(f"../../data/colet/participant{participant-1}/task{task-1}", exist_ok=True)
        gaze = pd.read_csv(f"../../data/colet/exports/participant{participant}_task{task}_gaze.csv")
        pupil = pd.read_csv(f"../../data/colet/exports/participant{participant}_task{task}_pupil.csv")
        blinks = pd.read_csv(f"../../data/colet/exports/participant{participant}_task{task}_blinks.csv")
        annot = pd.read_csv(f"../../data/colet/exports/participant{participant}_task{task}_annotation.csv").iloc[0].to_dict()
        
        gaze.to_csv(f"../../data/colet/participant{participant-1}/task{task-1}/gaze.csv", index=False)
        pupil.to_csv(f"../../data/colet/participant{participant-1}/task{task-1}/pupil.csv", index=False)
        blinks.to_csv(f"../../data/colet/participant{participant-1}/task{task-1}/blinks.csv", index=False)
        
        with open(f"../../data/colet/participant{participant-1}/task{task-1}/annotation.json", "w+") as f:
            json.dump(annot, f)
        print("participant", participant, "task", task, "done")
```
