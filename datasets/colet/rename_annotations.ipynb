```python
from glob import glob
from tqdm import tqdm
```

```python
files = glob("../../data/colet/*/*/annotation.json")
len(files)
```

```python
import json
for file in tqdm(files):
    with open(file, "r") as f:
        data = json.load(f)
    
    data = {
        k.split("_")[0]: v for k, v in data.items()
    }
    del data["mean"]
    
    with open(file, "w") as f:
        json.dump(data, f, indent=2)
```
